Histogram Equalization


```python

from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np


files.upload()

img = cv2.imread("your_image.jpg", cv2.IMREAD_GRAYSCALE)


height, width = img.shape[:2]
N = height * width


histogram, _ = np.histogram(img.flatten(), 256, [0, 256])


pdf = histogram / float(N)
cdf = np.cumsum(pdf)
cdf_normalized = np.uint8(255 * cdf)


high_contrast = cdf_normalized[img]


cv2.imwrite("high_contrast.jpg", high_contrast)


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(img.ravel(), 256, [0, 256])
plt.title("Original Histogram")
plt.xlabel("Intensity Values")
plt.ylabel("Pixel Count")

plt.subplot(1, 2, 2)
plt.hist(high_contrast.ravel(), 256, [0, 256])
plt.title("Equalized Histogram")
plt.xlabel("Intensity Values")
plt.ylabel("Pixel Count")

plt.show()


plt.imshow(high_contrast, cmap="gray")
plt.title("Histogram Equalized Image")
plt.axis("off")
plt.show()
```

Gradient Algorithm


```python

from google.colab import files
from PIL import Image
import cv2
import numpy as np


files.upload()

# Reading  in grayscale
img_gray = cv2.imread("your_image.jpg", cv2.IMREAD_GRAYSCALE)


h, w = img_gray.shape[:2]

# make gradient image
grad_img = np.zeros_like(img_gray)


for i in range(h):
    for j in range(w - 1):
        a = min(img_gray[i, j + 1], img_gray[i, j])
        temp_arr = abs(img_gray[i, j + 1] - img_gray[i, j])
        grad_img[i, j] = temp_arr


gradient_img = Image.fromarray(grad_img)
gradient_img.save("gradient.jpg")


import matplotlib.pyplot as plt
plt.imshow(grad_img, cmap="gray")
plt.title("Gradient Image")
plt.axis("off")
plt.show()
```

SHARPENING ALGORITHM

```python
import copy


files.upload()


im = cv2.imread("your_image.jpg", cv2.IMREAD_COLOR)


def sharpenFiltering(img, alpha=0.5): #adjust alpha for sharpening effect
    inputImg = copy.deepcopy(img.astype(np.float32))

    # to grayscale
    inputImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # black image of same size
    outputImg = np.zeros(inputImg.shape, dtype=np.float32)

    # Pad image with zeros
    inputImg = np.pad(inputImg, ((1, 1), (1, 1)), mode='constant', constant_values=0)

    # smoothening filters for less aggressive sharpening
    fh = np.array([[-0.5, -1.0, -0.5], [0.0, 0.0, 0.0], [0.5, 1.0, 0.5]], dtype=np.float32)
    fy = np.array([[-0.5, 0.0, 0.5], [-1.0, 0.0, 1.0], [-0.5, 0.0, 0.5]], dtype=np.float32)


    for row in range(1, inputImg.shape[0] - 1):
        for col in range(1, inputImg.shape[1] - 1):
            dx, dy = 0.0, 0.0

            # Convolving both filters
            for x_filter in range(3):
                for y_filter in range(3):
                    dx += inputImg[row + x_filter - 1, col + y_filter - 1] * fh[x_filter, y_filter]
                    dy += inputImg[row + x_filter - 1, col + y_filter - 1] * fy[x_filter, y_filter]

            # Compute magnitude of gradient
            pixel = np.sqrt(dx * dx + dy * dy)
            outputImg[row - 1, col - 1] = pixel

    # Normalization
    outputImg = np.clip((outputImg / np.max(outputImg)) * 255.0, 0, 255).astype(np.uint8)

    # Blend with original image to reduce the effect
    blended = cv2.addWeighted(inputImg[1:-1, 1:-1], 1 - alpha, outputImg, alpha, 0)

    return blended


output = sharpenFiltering(im, alpha=0.3) #keeping it 0.3 for reduced effect


cv2.imwrite("/content/sharpened_reduced.jpg", output)

# to RGB for visualization
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.axis("off")
plt.title("Original Image")
plt.imshow(im)

plt.subplot(1, 2, 2)
plt.axis("off")
plt.title("Reduced Sharpening Effect")
plt.imshow(output, cmap="gray")

plt.show()
```
